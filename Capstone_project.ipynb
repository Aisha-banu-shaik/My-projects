2)Apply an appropriate classification algorithm to build a model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
x= df.iloc[:,[0,1,2,3,4,5,6,7]].values
y= df.iloc[:,8].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
print(log_reg.score(x_train,y_train))
print(log_reg.score(x_test,y_test))
```

```python
print(classification_report(y,log_reg.predict(x)))
```

3)Compare various models with the results from KNN algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=7,metric='minkowski',p=2)
model.fit(x_train,y_train)
```

```python
y_pred=log_reg.predict(x_test)
accuracy_score(y_test,y_pred)*100
```

#Support Vector Classifier

```python
from sklearn.svm import SVC
model2=SVC(kernel='rbf',gamma='auto')
model2.fit(x_train,y_train)
```

```python
model2.score(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

#RandomForest classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model3=RandomForestClassifier(n_estimators=11)
model3.fit(x_train,y_train)
```

```python
model3.score(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```

#Decission Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model4.score(x_train,y_train)
```

```python
model4.score(x_test,y_test)
```

4)Create a classification report by analyzing sensitivity, specificity, AUC (ROC curve), etc.

#ROC Curve

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
probs = log_reg.predict_proba(x)
probs = probs[:, 1]
auc = roc_auc_score(y, probs)
fpr, tpr, thresholds = roc_curve(y, probs)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
```

#ROC curve defining specificity

```python
probs = model.predict_proba(x)
probs = probs[:, 1]
auc = roc_auc_score(y, probs)
print('AUC: %.3f' % auc)
fpr, tpr, thresholds = roc_curve(y, probs)
print("True Positive Rate - {}, False Positive Rate - {} Thresholds - {}".format(tpr,fpr,thresholds))
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
```

#Precision Recall Curve for Logistic Regression

```python
from sklearn.metrics import precision_recall_curve,f1_score,average_precision_score,auc
```

```python
probs = log_reg.predict_proba(x)
probs = probs[:, 1]
yhat = log_reg.predict(x)
precision, recall, thresholds = precision_recall_curve(y, probs)
f1 = f1_score(y, yhat)
auc = auc(recall, precision)
ap = average_precision_score(y, probs)
print('f1=%.3f auc=%.3f ap=%.3f' % (f1, auc, ap))
plt.plot([0, 1], [0.5, 0.5], linestyle='--')
plt.plot(recall, precision, marker='.')
```
